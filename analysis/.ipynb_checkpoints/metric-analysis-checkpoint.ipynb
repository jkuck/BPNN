```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv('./MaxMarginal_BP.csv')
print(data)
```

```python
cur_data = data.loc[
    (data['UPDATES']=='PARALL') &
    data['ATTRACTIVE_FIELD'] 
]
values = {
    idx: cur_data[idx].values
    for idx in cur_data.columns
}
values['MSE_DiffLog'] = values['RMSE_DiffLog']**2
metric_names = [
    'ACC_GraphState', 'ACC_VarState',
    'RMSE_LogScore', 'L1_LogScore',
    'CrossEntropy_Prob', 'RMSE_Prob', 'L1_Prob',
    'RMSE_DiffLog', 'L1_DiffLog',
]
loss_names = [
    'MSE_DiffLog', 'CrossEntropy_Prob',
]
METRIC_PLOT_NUM = 10
METRIC_LOSS_PLOT_NUM = 4

# analyze metrics
metric_corr = metric_correlations(values, metric_names)
print(metric_corr)
for row, col in np.nonzero(np.abs(metric_corr)>=(np.sort(np.abs(metric_corr).reshape(-1))[-METRIC_PLOT_NUM])):
    plot_correlations(values, metric_names[row], metric_names[col])
    
# analyze losses and metrics
metric_loss_corr = metric_loss_correlations(values, metric_names, loss_names)
print(metric_loss_corr)
for li, lname in enumerate(loss_names):
    cur_corr = np.abs(metric_loss_corr[li]).reshape(-1)
    for idx in np.nonzero(cur_corr>=(np.sort(cur_corr)[-METRIC_LOSS_PLOT_NUM])):
        plot_correlations(values, lname, metric_names[idx])
    
```
